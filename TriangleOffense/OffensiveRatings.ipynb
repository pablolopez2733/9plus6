```python
import pandas as pd
import numpy as np
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
from matplotlib import rcParams
from mpldatacursor import datacursor
import mplcursors
```

```python
#Leemos datos
df=pd.read_csv("LALSeasons.csv")
df.head()
```

```python
#Limpiamos columnas molestas
cols = []
for i in range (4,22):
    cols.append(i)

df.drop(df.columns[cols],axis=1,inplace=True)
df.head()
```

```python
#df_Ortg=df[['Temporada', 'ORtg']].copy()
#g = sns.relplot(x="Temporada", y="ORtg", data=df)
xlabels=df["Temporada"]
ax = sns.lineplot(x="Temporada", y="ORtg", data=df,marker='o',color='blueviolet')
ax.set_xticklabels(xlabels,rotation=90)
ax.set_title('ORtg por temporada de los Lakers')
sns.set_style("whitegrid")
```

```python
seasons=pd.read_csv("https://github.com/pablolopez2733/9plus6/raw/master/TriangleOffense/AdvancedStats99_16.csv")
seasons.head()
```

```python
leagueOrtg = seasons[['Season', 'Tm', 'ORtg']].copy()
leagueOrtg.head()
```

```python
seasons_avg=leagueOrtg.groupby("Season").mean().reset_index()
seasons_avg
```

```python
seasons_avg["LAL_ORtg"]=df["ORtg"]
seasons_avg.head()
```

```python
rcParams['figure.figsize'] = 8.4,6.46
xlabels=seasons_avg["Season"]
ax = sns.lineplot(x="Season", y="ORtg", data=seasons_avg,marker='o',color='blueviolet',label="Promedio de la liga")
ax = sns.lineplot(x="Season", y="LAL_ORtg", data=seasons_avg,marker='o',color='gold',label="Los Angeles Lakers")
ax.set_xticklabels(xlabels,rotation=90)
ax.set_title('ORtg por temporada ')
sns.set_style("whitegrid")
ax.legend()
```
